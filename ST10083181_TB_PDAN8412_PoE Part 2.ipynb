# What is logistic regression:
- According to Anon (n.d), Logistic regression, also referred to as Logit model, is a statistical modelling technique used to estimate the probability of a binary outcome based on one or more independent variables. Unlike linear regression, which predicts continuous outcomes, logistic regression maps the linear combination of input features through the logistic function, transforming the output into a value between 0 and 1. This is achieved by applying the logit transformation to the odds of the event occurring. 

- The dependent variable in logistic regression is categorical and dichotomous, meaning it fits into one of two categories (e.g., "yes" or "no," "pass" or "fail"). The dependent variable are the factors that may influence the outcome and can fall into several categories:
1.	Continuous Variables: These can take on any value within a range (e.g., temperature in degrees Celsius, weight in grams). Continuous data can be classified further into:
     -	Interval Data: Where the intervals between values are equal but there is no true zero (e.g., temperature in degrees Celsius).
     -	Ratio Data: Where intervals are equal and there is a meaningful zero point (e.g., weight in grams, where zero means no weight).
2.	Discrete Variables: These can take on specific values and are often counts of items (e.g., the number of customers visiting a store).
3.	Ordinal Variables: These can be categorized into a defined order or ranking (e.g., customer satisfaction ratings of "poor," "fair," "good," "excellent").

- The primary aim of logistic regression is to model the relationship between the dependent variable and the independent variables to facilitate predictions. By analysing historical data, logistic regression identifies patterns and relationships that can inform decisions in various fields, such as healthcare, finance, and marketing. For example, it can predict whether a patient has a certain disease based on various health metrics, forecast whether a customer will purchase a product based on their browsing behavior, or assess the likelihood of loan default based on financial history. 

- Anon (n.d) states that logistic regression is a versatile statistical modelling technique that can be adapted to various types of dependent variables and data structures. The primary types of logistic regression are:
     1. Binary Logistic Regression: This is the most common form of logistic regression, used when the dependent variable has two possible outcomes, typically coded as 0 and 1. E.g. Predicting whether a customer will default on a loan (yes/no).
     2. Multinomial Logistic Regression: This type is used when the dependent variable has more than two categories that are not ordered. It generalizes binary logistic regression to handle multiple classes. E.g. Classifying a type of fruit based on colour (e.g., apple, orange, banana).
     3. Ordinal Logistic Regression: This variation is used when the dependent variable is ordinal, meaning the categories have a natural order but are not equally spaced. E.g. Predicting customer satisfaction levels (e.g., dissatisfied, neutral, satisfied).
     4. Regularized Logistic Regression: Techniques such as Lasso (L1) or Ridge (L2) regularization can be applied to logistic regression to prevent overfitting, especially in high-dimensional datasets. E.g. Regularized logistic regression might be used when there are many predictor variables, such as in genomic studies.
     5. Hierarchical Logistic Regression: This type considers data that is grouped at multiple levels (e.g., individuals nested within organizations). It accounts for the correlation within clusters. E.g. Analysing patient outcomes across different hospitals, where patients are grouped by hospital.

- According to Anon (n.d), Logistic regression differs from linear regression and other general linear models that rely on ordinary least squares algorithms, as it does not impose many of the same key assumptions, particularly regarding linearity, normality, homoscedasticity, and measurement levels. The assumptions that still hold for logistic regression is that it does not require a linear relationship between the dependent and independent variables. The residuals or error terms do not need to follow a normal distribution. There is no requirement for homoscedasticity. Lastly, the dependent variable in logistic regression does not need to be measured on an interval or ratio scale.

- According to Thanda, A (2023), Logistic regression serves as a powerful tool, offering several advantages and disadvanatges such as:
     - Interpretability: The model provides clear coefficients, which can be exponentiated to yield odds ratios, making it easier to interpret the impact of independent variables on the likelihood of the outcome.
     - Flexibility: Logistic regression can accommodate various types of independent variables, including continuous, categorical, and binary predictors.
     - Assumption of Linearity: Logistic regression assumes a linear relationship between the independent variables and the log odds of the dependent variable. If the relationship is not linear, the model may not perform well, potentially leading to inaccurate predictions.
     - Multicollinearity: When independent variables are highly correlated, it can cause issues with the estimation of coefficients. Multicollinearity makes it difficult to determine the individual effect of each variable, leading to inflated standard errors and less reliable results.

- Binary Outcome: The primary variable of interest, default, is binary in nature (i.e., it indicates whether an individual has defaulted on their credit card payment or not). Logistic regression is specifically designed for situations where the outcome variable is categorical with two possible values (0 or 1, Yes or No).
- Probabilistic Interpretation: Logistic regression models the probability that the target variable belongs to a particular category (e.g., the probability of defaulting). This probabilistic output is essential for making informed decisions in risk assessment. 
2. Binary Outcome with Clear Predictor Variables
- Simplicity of Interpretation: The dataset features a binary outcome variable (default), which directly indicates whether an individual has defaulted on their credit card payments. This binary nature aligns perfectly with the foundational purpose of logistic regression, which is to model binary outcomes. Furthermore, the features in the dataset—such as balance, income, and student status—are well-defined predictor variables that influence the likelihood of defaulting.
- Balance: A higher account balance may suggest better financial health, potentially reducing the likelihood of default.
- Income: Higher income levels generally correlate with a greater ability to repay debts, making it a strong predictor of default risk.
- Student Status: Being a student could indicate a different risk profile compared to non-students, as students might have lower or unstable income levels.
3. Relationship Between Features and Outcome:
- Linear Relationships: Logistic regression assumes a linear relationship between the log-odds of the dependent variable and the independent variables.This means that the features (e.g., student, balance, income) can be linearly combined to estimate the log-odds of defaulting. In practical terms, it allows us to interpret the effect of each feature on the likelihood of default.
- Continuous and Categorical Features: The dataset contains a mix of continuous (e.g., balance, income) and categorical (e.g., student) variables. This interpretability is crucial for stakeholders (e.g., financial institutions) to understand the risk factors associated with defaults.
- Odds Ratios: The model allows us to calculate odds ratios, which help quantify the effect of a one-unit change in a predictor variable on the odds of the outcome occurring. This is particularly useful for decision-making in finance.
5. Assumptions of Logistic Regression
- Independence of Observations: The dataset comprises independent observations, meaning that each individual’s default status does not affect others. Logistic regression operates under this assumption, making it a suitable choice.
- Non-constant Variance: Unlike linear regression, which assumes constant variance of errors (homoscedasticity), logistic regression accounts for non-constant variance in the dependent variable, aligning with the nature of binary outcomes.

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np
from imblearn.over_sampling import SMOTE
from sklearn.utils import resample

# Data Preparation
def prepare_data(data):
    target_variable = 'default'
    features = data.drop(columns=[target_variable])
    X = features
    y = data[target_variable]

    # Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

    # Handle class imbalance using SMOTE
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    # Random Undersampling
    majority = X_train[y_train == 0]
    minority = X_train[y_train == 1]
    majority_undersampled = resample(majority, 
                                     replace=False,
                                     n_samples=len(minority), 
                                     random_state=42)

    X_train_balanced = np.vstack((majority_undersampled, minority))
    y_train_balanced = np.concatenate((np.zeros(len(majority_undersampled)), np.ones(len(minority))))

    # Scale features
    scaler = StandardScaler()
    X_train_resampled_scaled = pd.DataFrame(scaler.fit_transform(X_train_resampled), columns=X_train.columns)
    X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
    
    return X_train_resampled_scaled, X_test_scaled, y_train_resampled, y_test 

#Code attribution
# Kartik, T. 2024. ML | Handling Imbalanced Data with SMOTE and Near Miss Algorithm in Python. GeeksforGeeks [Online] Available at: https://seaborn.pydata.org/generated/seaborn.heatmap.html [Accessed 09 October 2024]
```

```python
# Import necessary libraries
import numpy as np
from sklearn.metrics import accuracy_score

# Logistic Regression Class with Regularization
class LogisticRegression:
    def __init__(self, learning_rate=0.0001, num_iterations=10000, regularization='l2', lambda_param=0.01):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.regularization = regularization
        self.lambda_param = lambda_param
        self.weights = None
        self.bias = None
        self.losses = []
        self.validation_accuracies = []  # List to store validation accuracies
        self.validation_losses = []        # List to store validation losses

    def sigmoid(self, z):
        z = np.clip(z, -500, 500)  # Avoid overflow issues
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y, X_val=None, y_val=None):
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0

        for i in range(self.num_iterations):
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = self.sigmoid(linear_model)

            dw = (1 / num_samples) * np.dot(X.T, (y_predicted - y))

            # Regularization
            if self.regularization == 'l2':
                dw += (self.lambda_param / num_samples) * self.weights  # L2 regularization
            elif self.regularization == 'l1':
                dw += (self.lambda_param / num_samples) * np.sign(self.weights)  # L1 regularization

            db = (1 / num_samples) * np.sum(y_predicted - y)

            # Update weights and bias
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
            
            # Calculate loss with regularization
            loss = -np.mean(y * np.log(y_predicted + 1e-15) + (1 - y) * np.log(1 - y_predicted + 1e-15))
            if self.regularization == 'l2':
                loss += (self.lambda_param / (2 * num_samples)) * np.sum(self.weights ** 2)
            elif self.regularization == 'l1':
                loss += (self.lambda_param / num_samples) * np.sum(np.abs(self.weights))
                
            self.losses.append(loss)

            # Validation metrics
            if X_val is not None and y_val is not None:
                val_predicted = self.predict(X_val)
                val_accuracy = accuracy_score(y_val, val_predicted)
                self.validation_accuracies.append(val_accuracy)
                
                val_loss = -np.mean(y_val * np.log(self.sigmoid(np.dot(X_val, self.weights) + self.bias) + 1e-15) +
                                    (1 - y_val) * np.log(1 - self.sigmoid(np.dot(X_val, self.weights) + self.bias) + 1e-15))
                self.validation_losses.append(val_loss)

            if i % 200 == 0:
                print(f"Iteration {i}: Loss = {loss:.4f}, Validation Accuracy = {val_accuracy if X_val is not None else 'N/A'}")

    def predict(self, X, threshold=0.5):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self.sigmoid(linear_model)
        return np.where(y_predicted >= threshold, 1, 0)

# Code attribution
# Koishik, A. 2023. Logistic Regression From Scratch. Medium. [Online] Available at: https://medium.com/@koushikkushal95/logistic-regression-from-scratch-dfb8527a4226 [Accessed 09 October 2024]
```

```python
# Hyperparameter Tuning with Validation 
def random_search(model_class, X_train, y_train, X_val, y_val, param_dist, n_iter=10):
    best_model = None
    best_score = float('-inf')
    best_params = None

    for i in range(n_iter):
        params = {key: np.random.choice(value) for key, value in param_dist.items()}
        model = model_class(**params)
        model.fit(X_train, y_train, X_val, y_val)

        # Evaluate on validation set
        score = accuracy_score(y_val, model.predict(X_val))
        
        # Print the current iteration results
        print(f"Iteration {i + 1}/{n_iter}: Params = {params}, Validation Score = {score:.4f}")

        if score > best_score:
            best_score = score
            best_model = model
            best_params = params

    print(f"Best Parameters: {best_params}, Best Score: {best_score:.4f}")
    return best_model, best_score

# Prepare Data
X_train_resampled_scaled, X_test_scaled, y_train_balanced, y_test = prepare_data(data)

# Split training data for validation
X_train, X_val, y_train, y_val = train_test_split(X_train_resampled_scaled, y_train_balanced, test_size=0.2, random_state=42)

# Hyperparameter search space
param_dist = {
    'learning_rate': [0.0001, 0.001, 0.01],
    'num_iterations': [10000, 15000, 20000],
    'regularization': ['l1', 'l2'],
    'lambda_param': [0.01, 0.1, 1.0]
}

# Run random search for hyperparameter tuning
best_model, best_score = random_search(LogisticRegression, X_train, y_train, X_val, y_val, param_dist)

# Code attribution
# Hassan, M. 2023. Tuning Model Hyperparameters With Random Search. Medium. [Online] Available at: https://medium.com/@hammad.ai/tuning-model-hyperparameters-with-random-search-f4c1cc88f528 [Accessed 10 October 2024]
```

# Receiver Operating Characteristic (ROC) Curve
- This graph is a Receiver Operating Characteristic (ROC) curve, which is used to evaluate the performance of a binary classification model. 

- Axes 
- False Positive Rate (FPR):
     - The x-axis represents the False Positive Rate, which quantifies how often negative instances (non-defaults) are incorrectly classified as positive (defaults). A lower FPR is desirable as it indicates fewer false alarms.
- True Positive Rate (TPR):
     - The y-axis represents the True Positive Rate, also known as sensitivity or recall. It indicates how effectively the model correctly identifies positive instances (defaults). A higher TPR is preferable, reflecting a model's ability to correctly classify defaults.

- ROC Curve:
     - The blue curve illustrates the trade-off between sensitivity (TPR) and specificity (1 - FPR) for various classification thresholds.
     - An ideal model would achieve a TPR of 1.0 while maintaining an FPR of 0.0, represented by a point at the top-left corner of the plot.
     - The curve's position above the diagonal line indicates that the model significantly outperforms a random classifier.

- Diagonal Line:
     - The red dashed line serves as a baseline representing the performance of a random classifier. Any model that falls below this line lacks predictive power, effectively guessing at random.

- Area Under the Curve (AUC)
     - The AUC value calculated at 0.9510 suggests that the model demonstrates excellent performance in distinguishing between the two classes. Specifically:

- AUC Interpretation:
     - 0.5: Indicates no discriminative power (random guessing).
     - 0.7 - 0.8: Indicates acceptable discrimination.
     - 0.8 - 0.9: Indicates excellent discrimination.
- 0.9: Indicates outstanding discrimination.
- An AUC of 0.9510 signifies that the model has a very high ability to correctly classify instances, which is particularly critical in domains such as finance where misclassification can lead to substantial consequences.

- Model Effectiveness
- High Discriminative Ability:
     - The ROC curve is consistently above the diagonal line, reflecting a robust ability to differentiate between defaults and non-defaults.
- Reliability:
     - An AUC of 0.9510 not only indicates model reliability but also instills confidence in its predictions, suggesting that the model is well-suited for the Logistic regression task.
